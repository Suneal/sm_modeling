```python
dictionary = {}

for i in expertises['fields']:
    if " / " in i:
        i = i.replace("/", "&")
    elif "/" in i:
        i = i.replace("/", " & ")
    if "and " in i:
        i = i.replace("and ", "& ")
    if ", " in i:
        i = i.replace(", ", " & ")
    if i.lower() == "computer science major":
        i = "computer science"
    if i.lower() == "undergraduate computer science":
        i = "computer science"
    if i.lower() == "computer science & music double major":
        i = "computer science & music"
    if i.lower() not in dictionary:
        dictionary[i.lower()] = 1
    else:
        dictionary[i.lower()] += 1

pprint.pprint(dictionary)
```

```python
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        axs.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')
```

```python
# Interfaces to use
combined = pd.read_csv("Combined_tony.csv")
combined = combined['Q5'][2:]
interface_count = {}
for item in combined:
    list_of_item = item.split(";")
    for i in list_of_item:
#         print(f"i = {i}")
        if i not in interface_count:
            interface_count[i] = 1
        else:
            interface_count[i] += 1


sorted_interface_count = sorted(interface_count.items(), key=lambda kv: kv[1], reverse=True)
sorted_interface_count

new_interface_count = {}
for i in range( len(sorted_interface_count)):
    new_interface_count[sorted_interface_count[i][0]] = sorted_interface_count[i][1]

# new_interface_count
fig, axs = plt.subplots(figsize=(25,8))
plt.xticks(rotation=45, ha="center")
plt.bar(new_interface_count.keys(), new_interface_count.values(), width = 1)
```
