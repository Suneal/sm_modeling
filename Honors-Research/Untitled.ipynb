```python
trigger_var = ['awayclose','away', 'lesser-home', 'greater-lesser-on']
trigger_dev = ['null && switch','null', 'light && null', 'light && temperature && switch']
trigger_cap = ['locationMode && motion','locationMode', 'temperature && locationMode', 'light && Temperature && light']
```

```python
for i in range(len(trigger_var)):
    
    list_of_var = trigger_var[i].split("-")
    list_of_dev = trigger_dev[i].split(" && ")
    list_of_cap = trigger_cap[i].split(" && ")
    
    if (len(list_of_var) != len(list_of_dev)):
        print("########################################################")
        print(f"ERROR at index = {i}, len(list_of_var) != list_of_dev ")
        print("########################################################")
        continue
    if (len(list_of_var) != len(list_of_cap)):
        print("########################################################")
        print(f"ERROR at index = {i}, len(list_of_var) != list_of_cap ")
        print("########################################################")
        continue
    if (len(list_of_cap) != len(list_of_dev)):
        print("########################################################")
        print(f"ERROR at index = {i}, len(list_of_cap) != list_of_dev ")
        print("########################################################")
        continue
    
    if '-' in trigger_var[i]:
        list_of_i = trigger_var[i].split("-")        
        modification = sorted(range(len(list_of_i)),key=list_of_i.__getitem__) 

        list_of_dev = trigger_dev[i].split(" && ")

        temp_dev = copy.deepcopy(list_of_dev)
        for j in range (len(modification)):
            temp_dev[j] = list_of_dev[modification[j]]

        list_of_cap = trigger_cap[i].split(" && ")
        temp_cap = copy.deepcopy(list_of_cap)
        for k in range (len(modification)):
            temp_cap[k] = list_of_cap[modification[k]]
                
        
```

```python
final = pd.read_csv("Training Set - Final V1 - Sheet1.csv")
# final.head()
```

```python
trigger_devices = final["Trigger Devices Copy"]
# trigger_devices
```

```python
trigger_variables = final["Trigger variables copy"]
# trigger_variables
```

```python
trigger_capabilities = final["Trigger Capabilities"]
# trigger_capabilities
```

```python
final_trigger_var = []
final_trigger_dev = []
final_trigger_cap = []
for i in range(len(trigger_capabilities)):
    if isinstance(trigger_variables[i],float):
        trigger_variables[i] = "null"
    if isinstance(trigger_devices[i],float):
        trigger_devices[i] = "null"
    if isinstance(trigger_capabilities[i],float):
        trigger_capabilities[i] = "null"
    list_of_var = trigger_variables[i].split("-")
    list_of_dev = trigger_devices[i].split(" && ")
    list_of_cap = trigger_capabilities[i].split(" && ")

    if (len(list_of_var) != len(list_of_dev)):
        print("########################################################")
        print(f"ERROR at index = {i}, len(list_of_var) != list_of_dev ")
        print()
        print("########################################################")
        continue
    if (len(list_of_var) != len(list_of_cap)):
        print("########################################################")
        print(f"ERROR at index = {i}, len(list_of_var) != list_of_cap ")
        print("########################################################")
        continue
    if (len(list_of_cap) != len(list_of_dev)):
        print("########################################################")
        print(f"ERROR at index = {i}, len(list_of_cap) != list_of_dev ")
        print("########################################################")
        continue
    if '-' in trigger_variables[i]:
        
        list_of_dev = trigger_devices[i].split(" && ")        
        modification = sorted(range(len(list_of_dev)),key=list_of_dev.__getitem__) 
        temp_dev = sorted(list_of_dev)
        final_trigger_dev.append(' && '.join(temp_dev))
        
        list_of_var = trigger_variables[i].split("-")
        temp_var = copy.deepcopy(list_of_var)
        for j in range (len(modification)):
            temp_var[j] = list_of_var[modification[j]]

        final_trigger_var.append('-'.join(temp_var))
        
#         final_trigger_dev.append(' && '.join(temp_dev))
        list_of_cap = trigger_capabilities[i].split(" && ")
        temp_cap = copy.deepcopy(list_of_cap)
        for k in range (len(modification)):
            temp_cap[k] = list_of_cap[modification[k]]
        
        final_trigger_cap.append(' && '.join(temp_cap))
    else:
        final_trigger_var.append(trigger_variables[i])
        final_trigger_dev.append(trigger_devices[i])
        final_trigger_cap.append(trigger_capabilities[i])
```

```python
action_devices = final["Action Devices"]
action_variables = final["Action Variables Copy"]
action_capabilities = final["Action Capabilities"]

final_action_var = []
final_action_dev = []
final_action_cap = []
for i in range(len(action_capabilities)):
    if isinstance(action_variables[i],float):
        action_variables[i] = "null"
    if isinstance(action_devices[i],float):
        action_devices[i] = "null"
    if isinstance(action_capabilities[i],float):
        action_capabilities[i] = "null"
    list_of_var = action_variables[i].split("-")
    list_of_dev = action_devices[i].split(" && ")
    list_of_cap = action_capabilities[i].split(" && ")

    if (len(list_of_var) != len(list_of_dev)):
        print("########################################################")
        print(f"ERROR at index = {i}, len(list_of_var) != list_of_dev ")
        print()
        print("########################################################")
        continue
    if (len(list_of_var) != len(list_of_cap)):
        print("########################################################")
        print(f"ERROR at index = {i}, len(list_of_var) != list_of_cap ")
        print("########################################################")
        continue
    if (len(list_of_cap) != len(list_of_dev)):
        print("########################################################")
        print(f"ERROR at index = {i}, len(list_of_cap) != list_of_dev ")
        print("########################################################")
        continue
    if '-' in action_variables[i]:
        
        list_of_dev = action_devices[i].split(" && ")        
        modification = sorted(range(len(list_of_dev)),key=list_of_dev.__getitem__) 
        temp_dev = sorted(list_of_dev)
        final_action_dev.append(' && '.join(temp_dev))
        
        list_of_var = action_variables[i].split("-")
        temp_var = copy.deepcopy(list_of_var)
        for j in range (len(modification)):
            temp_var[j] = list_of_var[modification[j]]

        final_action_var.append('-'.join(temp_var))

        list_of_cap = action_capabilities[i].split(" && ")
        temp_cap = copy.deepcopy(list_of_cap)
        for k in range (len(modification)):
            temp_cap[k] = list_of_cap[modification[k]]
        
        final_action_cap.append(' && '.join(temp_cap))
    else:
        final_action_var.append(action_variables[i])
        final_action_dev.append(action_devices[i])
        final_action_cap.append(action_capabilities[i])
```
