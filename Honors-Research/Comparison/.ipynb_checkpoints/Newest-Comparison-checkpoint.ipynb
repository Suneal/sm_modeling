```python
# Getting the trigger device user id in the form of a dictionary with key as device and value as 
trigger_device_user_dict = {}

for i in range(len(new_csv['Trigger Devices'].dropna())):
    trigger = new_csv['Trigger Devices'][i]
    if isinstance(trigger, float):
        trigger = "Mode"
    if "&&" in trigger:
        list_of_trigger = trigger.split(" && ")
        for item in list_of_trigger:
            if item == "null":
                item = "Mode"
            if item not in trigger_device_user_dict:
                trigger_device_user_dict[item] = set([new_csv['UserId'][i]])
            else:
                trigger_device_user_dict[item].add(new_csv['UserId'][i])
    else:
        if trigger not in trigger_device_user_dict:
            trigger_device_user_dict[trigger] = set([new_csv['UserId'][i]])
        else:
            trigger_device_user_dict[trigger].add(new_csv['UserId'][i])

# trigger_device_user_dict
```

```python
# Getting the action device id in the form of a dictionary
# Getting the trigger device user id in the form of a dictionary with key as device and value as 
action_device_user_dict = {}

for i in range(len(new_csv['Action Devices'])):
    action = new_csv['Action Devices'][i]
    if isinstance(action, float):
        action = "Mode"
    if "&&" in action:
        list_of_action = action.split(" && ")
        for item in list_of_action:
            if item == "null" or item == "nan" or isinstance(action,float):
                item = "Mode"
            if item == "Light":
                item = "Light Bulb"
            if item == "AirPurifier":
                item = "Air Purifier"
            if item not in action_device_user_dict:
                action_device_user_dict[item] = set([new_csv['UserId'][i]])
            else:
                action_device_user_dict[item].add(new_csv['UserId'][i])
    else:
        if action not in action_device_user_dict:
            action_device_user_dict[action] = set([new_csv['UserId'][i]])
        else:
            action_device_user_dict[action].add(new_csv['UserId'][i])
# action_device_user_dict
```

```python
total_device = {}
total_device_count = {}

for i in action_device_user_dict:
    total_device[i] = action_device_user_dict[i]

for i in trigger_device_user_dict:
    if i not in total_device:
        total_device[i] = trigger_device_user_dict[i]
    else:
        for j in trigger_device_user_dict[i]:
            total_device[i].add(j)
for i in total_device:
    total_device_count[i] = len(total_device[i])

sorted_total_device_count = sorted(total_device_count.items(),key=operator.itemgetter(1), reverse=True)
sorted_total_device_count

new_total_device_count = {}
for i in range( len(sorted_total_device_count)):
    new_total_device_count[sorted_total_device_count[i][0]] = sorted_total_device_count[i][1]

# new_total_device_count
```
