```python
# Imports
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import string
import seaborn as sns
import pprint
import pylab as pl
from pylab import rcParams
from collections import Counter
import operator
```

```python
old_csv = pd.read_csv("train-consistent.csv")
old_csv.head()
```

```python
old_csv.shape
```

```python
common_app = old_csv['Common app name']
intersect  = old_csv['Intersection: 0 = common; 1 = unique']
```

```python
common_app_dict = {}

for i in common_app:
    if type(i) is float:
        continue
    else:
        if i not in common_app_dict:
            common_app_dict[i] = 1
        else:
            common_app_dict[i] += 1
```

```python
pprint.pprint(common_app_dict)
```

```python
sorted_common_app_dict = sorted(common_app_dict.items(), key=lambda kv: kv[1],reverse=True)
```
